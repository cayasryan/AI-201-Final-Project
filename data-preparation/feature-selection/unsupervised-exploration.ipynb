```python
import pandas as pd
from sklearn.model_selection import train_test_split
import numpy as np
```

```python
df = pd.read_parquet('../../data/features/featureswithouttags.parquet')
df.columns = df.columns.droplevel(0)
df
```

```python
features = ['acousticness',	'danceability',	'energy',	'instrumentalness',	'liveness',	'speechiness',	'tempo',	'valence']
samples = df[features]
samples.shape
```

```python
from sklearn.cluster import KMeans
X = samples
#X_train, X_test = train_test_split(X)

kmeans = KMeans(n_clusters=6, random_state=0, n_init="auto").fit(X)
kmeans.labels_.shape
```

```python
songs_w_tag_cluster = pd.read_parquet('../../data/features/featuresv4_w_tagclusters.parquet')
songs_w_tag_cluster.columns = songs_w_tag_cluster.columns.droplevel(0)
songs_w_tag_cluster.head()
```

```python
Sample_val = songs_w_tag_cluster.sort_values(by=['tag_cluster'],ascending=True)
features = ['','acousticness',	'danceability',	'energy',	'instrumentalness',	'liveness',	'speechiness',	'tempo',	'valence', 'tag_cluster'] 
Sample_val = Sample_val[features]
Sample_val = Sample_val.drop_duplicates()
Sample_val
```

```python
X_val = Sample_val.drop(columns=['','tag_cluster'])
y_val = Sample_val['tag_cluster']
track_id = Sample_val['']
```

```python
y_pred = kmeans.predict(X_val)
```

```python
comp = np.stack((y_val.to_numpy(),y_pred),axis=1)
comp
```

```python
y_pred_df = pd.DataFrame(data=comp,columns=['ground_truth','predicted_cluster'])
```

```python
for i in range(1,7):
    cluster = y_pred_df[y_pred_df['ground_truth']==i]
    print(cluster['predicted_cluster'].value_counts())
```
