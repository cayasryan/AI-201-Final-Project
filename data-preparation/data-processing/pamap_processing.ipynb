```python
import numpy as np
import pandas as pd
import os
```

```python
# Do the processing for all dat files
directory1 = '../../data/raw/pamap2_data/Protocol'
directory2 = '../../data/raw/pamap2_data/Optional'
```

```python
combined_results = pd.DataFrame()
# combined_results_var = pd.DataFrame()

for filename in os.listdir(directory1):
    filepath = os.path.join(directory1, filename)

    subject_df = pd.read_csv(filepath, delimiter=' ', header = None)
    # subject_df = subject_df[top_feats]
    subject_number = int(filename.split('subject')[1].split('.dat')[0])
    subject_df['Subject'] = subject_number

    subject_df[0] = pd.to_datetime(subject_df[0], unit='s')
    # subject_df['Minute'] = subject_df[0].dt.floor('T')
    subject_df['Second'] = subject_df[0].dt.floor('S')
    subject_df = subject_df[subject_df[1] != 0] # remove activity ID = 0
    subject_df['ActivityGroup'] = (subject_df[1] != subject_df[1].shift()).cumsum()


    subject_df = subject_df.groupby(['Subject', 'ActivityGroup', 'Second',1]).mean().reset_index()
    # subject_df_var = subject_df.groupby(['Subject', 'ActivityGroup', 'Second',1]).var().reset_index()

    subject_df = subject_df.drop(columns=[0,'ActivityGroup']) # drop timestamp and activity group
    # subject_df_var = subject_df_var.drop(columns=[0,'ActivityGroup']) # drop timestamp and activity group
    

    combined_results = pd.concat([combined_results, subject_df], ignore_index=True)
    # combined_results_var = pd.concat([combined_results_var, subject_df_var], ignore_index=True)


    
```

```python
for filename in os.listdir(directory2):
    filepath = os.path.join(directory2, filename)

    subject_df = pd.read_csv(filepath, delimiter=' ', header = None)
    # subject_df = subject_df[top_feats]
    subject_number = int(filename.split('subject')[1].split('.dat')[0])
    subject_df['Subject'] = subject_number

    subject_df[0] = pd.to_datetime(subject_df[0], unit='s')
    # subject_df['Minute'] = subject_df[0].dt.floor('T')
    subject_df['Second'] = subject_df[0].dt.floor('S')
    subject_df = subject_df[subject_df[1] != 0] # remove activity ID = 0
    subject_df['ActivityGroup'] = (subject_df[1] != subject_df[1].shift()).cumsum()


    subject_df = subject_df.groupby(['Subject', 'ActivityGroup', 'Second',1]).mean().reset_index()
    # subject_df_var = subject_df.groupby(['Subject', 'ActivityGroup', 'Second',1]).var().reset_index()

    subject_df = subject_df.drop(columns=[0,'ActivityGroup']) # drop timestamp and activity group
    # subject_df_var = subject_df_var.drop(columns=[0,'ActivityGroup']) # drop timestamp and activity group
    

    combined_results = pd.concat([combined_results, subject_df], ignore_index=True)
    # combined_results_var = pd.concat([combined_results_var, subject_df_var], ignore_index=True)
```
